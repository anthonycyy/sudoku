```python
def is_zero(board):
    for i in range(9):
        for j in range(9):
            if board[i][j]==0:
                return i,j
    return None
```

```python
def is_valid(board,num,row,col):
    for j in range(9):
        if board[row][j]==num:
            return False
    for i in range(9):
        if board[i][col]==num:
            return False
    grid_x=row//3
    grid_y=col//3
    for i in range(3):
        for j in range(3):
            if board[3*grid_x+i][3*grid_y+j]==num:
                return False
    return True
```

```python
def print_board(board):
    for i in range(9):
        row_=""
        for j in range(9):
            if j%3==0:
                row_+=" "
            row_+=str(board[i][j])+" "
            if j%3==2 and j!=8:
                row_+="|"
        print(row_)
        if i%3==2 and i!=8:
            print(" "+"-"*21)
```

```python
def solve(board):
    print_board(board)
    print("")
    if is_zero(board) is None:
        print_board(board)
        return True
    else:
        row,col=is_zero(board)
    for num in range(1,10):
        if is_valid(board,num,row,col):
            board[row][col]=num
            print("Try "+str(num)+" at "+str(row)+str(col))
            if solve(board):
                return True
            #print("Backtrack at "+str(row)+str(col))
            board[row][col]=0
            print("Backtrack at "+str(row)+str(col))
    return False
```

```python
solve(board)
```
